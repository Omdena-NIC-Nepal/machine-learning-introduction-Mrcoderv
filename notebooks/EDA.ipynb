### Visualize the relationships between each feature in the dataset and the target variable (medv).
#### This will help identify potential correlations or patterns that could be useful for predictive modeling.

```python
plt.figure(figsize=(16, 12))
plt.style.use('seaborn-v0_8-darkgrid')  # Use a dark grid style for better aesthetics

# Use a color palette for scatter points
colors = sns.color_palette('coolwarm', len(df.columns[:-1]))

for i, col in enumerate(df.columns[:-1], 1):  # Exclude 'medv'
    plt.subplot(4, 4, i)
    plt.scatter(df[col], df['medv'], alpha=0.7, edgecolor='k', color=colors[i-1])
    plt.title(f'{col} vs MEDV', fontsize=12, fontweight='bold')
    plt.xlabel(col, fontsize=10)
    plt.ylabel('MEDV', fontsize=10)
    plt.xticks(fontsize=9)
    plt.yticks(fontsize=9)
    plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.suptitle('Feature Relationships with MEDV', fontsize=18, fontweight='bold', y=1.02)
plt.show()
```

# Correlation Matrix

The correlation matrix provides a visual representation of the relationships between all numerical features in the dataset.

```python
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix', fontsize=16, fontweight='bold', pad=20)
plt.xticks(fontsize=12, rotation=45, ha='right')
plt.yticks(fontsize=12, rotation=0)
plt.tight_layout()
plt.show()
```
